```python
import pandas as pd
import numpy as np

df = pd.read_csv('heart_disease.csv')
df
```

```python
# compute frequency distribution of a categorical variable
freq_dist=df['age'].value_counts()
print("Frequency Distribution:")
print(freq_dist)
```

```python
#Compute mean, median,and mode of a numerical variable
import scipy.stats as stats
numerical_var = df['age'].values
mean = np.mean(numerical_var)
median = np.median(numerical_var)
mode = stats.mode(numerical_var).mode[0]
print("Mean:",mean)
print("Median:",median)
print("Mode:",mode)
```

```python
#Compute range ,variance, and standard deviation of a numerical variable
range_val = np.max(numerical_var)
variance = np.var(numerical_var)
std_dev = np.std(numerical_var)
print("Range:",range_val)
print("Variance:",variance)
print("Standard Deviation:", std_dev)
```
